# Unemployment Analysis with Python

**Objective:** Perform exploratory data analysis on unemployment data to uncover regional and temporal trends, with a focus on the impact of the COVID-19 pandemic on unemployment rates in India.

**Tech Stack:** Python, pandas, matplotlib, seaborn, Jupyter Notebook

**Dataset:** "Unemployment in India" from Kaggle (search this exact title, publicly available, free account required to download).

> **Before running this notebook:** download the dataset CSV from Kaggle and place it in the same folder as this notebook. Update the filename in the cell below if it differs.


## Step 1: Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

## Step 2: Load the Dataset

```
df = pd.read_csv('Unemployment_in_India.csv')

# The Kaggle CSV often has leading/trailing spaces in column names - clean them up
df.columns = df.columns.str.strip()

df.head()
```

```
df.shape
```

## Step 3: Data Inspection and Cleaning

```
print("Column names:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```
# Drop any fully empty rows (common in this dataset)
df = df.dropna(how='all')

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'].str.strip(), dayfirst=True)

# Strip whitespace from text columns
text_cols = df.select_dtypes(include='object').columns
for c in text_cols:
    df[c] = df[c].str.strip()

df.info()
```

## Step 4: Exploratory Data Analysis

```
region_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)
region_avg
```

```
df['Month'] = df['Date'].dt.month
month_avg = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()
month_avg
```

**Observations:** The region-wise averages above show which states have historically had the highest unemployment rates. The month-wise averages help identify whether there is any seasonal pattern to unemployment independent of year-specific events like COVID-19.

## Step 5: Time-Series Line Chart (at least 3 states)

```
states_to_plot = ['Andhra Pradesh', 'Bihar', 'Delhi']  # update with states present in your dataset

plt.figure(figsize=(12, 6))
for state in states_to_plot:
    subset = df[df['Region'] == state].sort_values('Date')
    plt.plot(subset['Date'], subset['Estimated Unemployment Rate (%)'], marker='o', label=state)

plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate Over Time by State')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Observations:** This chart shows how unemployment trended over time for the selected states. Watch for a sharp spike around March-April 2020, which aligns with the nationwide COVID-19 lockdown.

## Step 6: Bar Chart - Top 10 States by Average Unemployment Rate

```
top10 = region_avg.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top10.values, y=top10.index, palette='rocket')
plt.xlabel('Average Unemployment Rate (%)')
plt.ylabel('State/Region')
plt.title('Top 10 States by Average Unemployment Rate')
plt.tight_layout()
plt.show()
```

**Observations:** These are the states with the highest average unemployment rate across the full time period in the dataset. Note whether these are consistently high (structural unemployment) or driven mainly by the COVID-19 spike months.

## Step 7: Correlation Heatmap

```
cols_of_interest = [
    'Estimated Unemployment Rate (%)',
    'Estimated Employed',
    'Estimated Labour Participation Rate (%)'
]

corr = df[cols_of_interest].corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation: Unemployment vs Employment vs Labour Participation')
plt.tight_layout()
plt.show()
```

**Observations:** Look at the sign and strength of each correlation. Typically, unemployment rate and estimated employed count show a negative relationship, while the link with labour participation rate can vary by region and time period.

## Step 8: Pre-COVID vs Post-COVID Comparison

```
covid_cutoff = '2020-03-01'

pre_covid = df[df['Date'] < covid_cutoff]
post_covid = df[df['Date'] >= covid_cutoff]

pre_mean = pre_covid['Estimated Unemployment Rate (%)'].mean()
post_mean = post_covid['Estimated Unemployment Rate (%)'].mean()

print(f"Pre-COVID average unemployment rate: {pre_mean:.2f}%")
print(f"Post-COVID average unemployment rate: {post_mean:.2f}%")
print(f"Change: {post_mean - pre_mean:+.2f} percentage points")
```

```
comparison = pd.DataFrame({
    'Period': ['Pre-COVID', 'Post-COVID'],
    'Average Unemployment Rate (%)': [pre_mean, post_mean]
})

plt.figure(figsize=(6, 5))
sns.barplot(x='Period', y='Average Unemployment Rate (%)', data=comparison, palette='Set2')
plt.title('Pre-COVID vs Post-COVID Average Unemployment Rate')
plt.tight_layout()
plt.show()
```

**Observations:** The comparison above quantifies the pandemic's impact on unemployment nationally.
